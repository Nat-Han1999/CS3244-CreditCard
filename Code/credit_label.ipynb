```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

application = pd.read_csv("application_record.csv", encoding = 'utf-8')
record = pd.read_csv("credit_record.csv", encoding = 'utf-8')
```

```python
begin_month = pd.DataFrame(record.groupby(["ID"])["MONTHS_BALANCE"].aggregate("min")) #find the minimum value of each id(begin month)
begin_month  = begin_month.rename(columns = {"MONTHS_BALANCE":"BEGIN_MONTH"})#change column name
begin_month
```

```python
reserve = begin_month[begin_month["BEGIN_MONTH"] <=-12 ] #calculate reserve number
reserve.info()
```

```python
new_record = record.merge(begin_month, how = "left", on="ID") #left join
new_record.iloc[8905:8916]
```

```python
new_record['STATUS'].value_counts()
```

```python
#calculate reserve number
new_record["TYPE"] = 0
# new_record["TYPE"][record['STATUS'] =='1']=1
new_record["TYPE"][record['STATUS'] =='2']=1
new_record["TYPE"][record['STATUS'] =='3']=1
new_record["TYPE"][record['STATUS'] =='4']=1
new_record["TYPE"][record['STATUS'] =='5']=1 
new_record.iloc[8905:8916]
```

```python
labels= pd.DataFrame(new_record.groupby(["ID"])["TYPE"].aggregate("max"))
labels  = labels.rename(columns = {"TYPE":"REALTYPE"})
```

```python
labels[labels['REALTYPE'] == 1]
```

```python
labels.value_counts() #if change to over 1 month, then bad customers will increase to 5000
```

```python
label_record = new_record.merge(labels, how = "left", on="ID")
label_record
```
