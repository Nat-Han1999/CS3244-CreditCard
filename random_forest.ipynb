# **Split features and labels, training and testing set**

```python
labels = data['REALTYPE']
data= data.drop('REALTYPE', axis = 1)
data= data.drop('ID', axis = 1)
data= data.drop('CNT_FAM_MEMBERS', axis = 1)
data = data.drop('Unnamed: 0', axis = 1)
# data = np.array(data)
```

# **Not handling imbalanced dataset(high accuracy but low f1)**

```python
np.random.seed(seed=0)  #to get the same answer
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(data, labels, test_size = 0.25)
print('Training Features Shape:', X_train.shape)
print('Training Labels Shape:', Y_train.shape)
print('Testing Features Shape:', X_test.shape)
print('Testing Labels Shape:', Y_test.shape)
#imbalanced dataset, other sampling method?
```

```python
print("Number of Good customers in train",list(Y_train).count(0))
print("Number of bad customers in train", list(Y_train).count(1))
print("Number of Good customers in test",list(Y_test).count(0))
print("Number of bad customers in test", list(Y_test).count(1))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators = 100)
rf_model = rf.fit(X_train, Y_train);
#what is the best number of estimators?
```

```python
from sklearn.metrics import f1_score, accuracy_score, confusion_matrix, roc_auc_score, roc_curve,recall_score, precision_score
predictions = rf.predict(X_test)
print(rf_model.score(X_train, Y_train))
print(rf_model.score(X_test, Y_test))
```

```python

f = f1_score(y_true = Y_test , y_pred = predictions)

print("F1-score:",f)
print("Accuracy",accuracy_score(y_true = Y_test , y_pred = predictions))
print("Recall:", recall_score(y_true = Y_test , y_pred = predictions))
print("Precision: ",precision_score (y_true = Y_test , y_pred = predictions))
print("Confusiont Matrix \n",confusion_matrix(y_true = Y_test , y_pred = predictions) )
tn, fp, fn, tp = confusion_matrix(y_true = Y_test , y_pred = predictions).ravel()
print("tn: ", tn)
print("fp: ", fp)
print("fn: ",fn )
print("tp: ", tp)
```

```python
from sklearn.inspection import permutation_importance
from matplotlib import pyplot as plt
rf_model.feature_importances_
plt.figure(figsize=(15, 15), dpi=80)
plt.barh(list(X_train.columns), rf_model.feature_importances_)
```
